```python
import matplotlib.pyplot as plt
import numpy as np
from IPython import display

%matplotlib inline
plt.style.use('ggplot')

x = np.arange(0,2*np.pi,0.02*np.pi)

for k in range(100):
    plt.clf()
    y = np.sin(x+k/10.0)
    plt.plot(x,y)
    plt.axis([0,6.3,-1.1,1.1])
    display.display(plt.gcf())
    display.clear_output(wait=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
import math

num_individuals = 100
num_steps = 100
speed = 0.01  # Adjust this value for the desired speed of movement
boundary = 1.0  # Size of the square area

x_coords = np.random.rand(num_individuals)
y_coords = np.random.rand(num_individuals)
headings = np.random.uniform(0, 2 * math.pi, num_individuals)

def update_positions(i):
    global x_coords, y_coords, headings
    
    # Update positions based on headings and speed
    x_coords += speed * np.cos(headings)
    y_coords += speed * np.sin(headings)

    # Wrap around the boundaries
    x_coords = np.where(x_coords > boundary, x_coords - boundary, x_coords)
    y_coords = np.where(y_coords > boundary, y_coords - boundary, y_coords)
    x_coords = np.where(x_coords < 0, x_coords + boundary, x_coords)
    y_coords = np.where(y_coords < 0, y_coords + boundary, y_coords)

    # Update headings slightly to simulate random changes
    headings += np.random.uniform(-0.1, 0.1, num_individuals)
    
    # Clear the previous frame and plot the new positions
    plt.clf()
    plt.scatter(x_coords, y_coords, marker='o', s=10)
    plt.xlim(0, boundary)
    plt.ylim(0, boundary)

fig, ax = plt.subplots()
ani = FuncAnimation(fig, update_positions, frames=num_steps, interval=100)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from IPython.display import display, clear_output
import time

num_individuals = 100
num_steps = 100
speed = 0.01  # Can be adjusted for the desired speed of movement
boundary = 1.0  # Size of the square area

x_coords = np.random.rand(num_individuals)
y_coords = np.random.rand(num_individuals)
headings = np.random.uniform(0, 2 * math.pi, num_individuals)

fig, ax = plt.subplots()

for i in range(num_steps):
    # Update positions based on headings and speed
    x_coords += speed * np.cos(headings)
    y_coords += speed * np.sin(headings)

    # Wrap around the boundaries
    x_coords = np.where(x_coords > boundary, x_coords - boundary, x_coords)
    y_coords = np.where(y_coords > boundary, y_coords - boundary, y_coords)
    x_coords = np.where(x_coords < 0, x_coords + boundary, x_coords)
    y_coords = np.where(y_coords < 0, y_coords + boundary, y_coords)

    # Update headings slightly to simulate random changes
    headings += np.random.uniform(-0.1, 0.1, num_individuals)
    
    # Clear the previous frame and plot the new positions
    plt.clf()
    plt.scatter(x_coords, y_coords, marker='o', s=10)
    plt.xlim(0, boundary)
    plt.ylim(0, boundary)
    
    # Display the current frame
    display(fig)
    clear_output(wait=True)
    
    # Pause briefly to control the animation speed
    time.sleep(0.1)

plt.close()  # Close the figure after the animation is done
```
